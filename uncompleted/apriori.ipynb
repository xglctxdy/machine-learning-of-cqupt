```python
import pandas as pd
import numpy as np

path = "../data/groceries - groceries.csv"
dataset = pd.read_csv(path)
print(dataset.head())
```

```python
# key为当前物品名字，值为其映射的数字
goods_map = {}
# key为物品编号，值为其数量
goods_num = {}
# 当前候选集合，以frozenset形式保存
cur_set = []
# 数据集，里面存的是每条数据的set形式
goods_set = []
# 被抛弃过的候选集，其超集也一定不是频繁集，里面存的是frozenset
dropout_set = []
```

```python
def create_goods(show=False):
    substitute = 0
    for i in range(len(dataset)):
        data = dataset.iloc[i]
        for index, element in enumerate(data):
            if index == 0:
                continue
            if pd.isnull(element):
                break
            if element not in goods_map:
                goods_map[element] = substitute
                goods_num[substitute] = 1
                substitute += 1
            else:
                goods_num[goods_map[element]] += 1
    if show:
        total_num = 0
        print("当前有", len(goods_map), "种商品，其名称和数量分别是")
        for key in goods_map.keys():
            print(key, goods_num[goods_map[key]])
            total_num += goods_num[goods_map[key]]
        print("一共", total_num, "件")


def data_to_frozenset(data):
    ans = []
    for index, element in enumerate(data):
        if index == 0:
            continue
        if pd.isnull(element):
            break
        ans.append(goods_map[element])
    return frozenset(ans)


def data_to_set(data):
    ans = []
    for index, element in enumerate(data):
        if index == 0:
            continue
        if pd.isnull(element):
            break
        ans.append(goods_map[element])
    return set(ans)


def merge_frozenset(set1, set2):
    return set1.union(set2)
```
